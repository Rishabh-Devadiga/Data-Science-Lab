```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.amazon.in/Imsa-Moda-Polycotton-Loose-Comfortable/dp/B0GWVFLGFD/?_encoding=UTF8&pd_rd_w=CUt6T&content-id=amzn1.sym.340182bc-8d5c-49c7-8b69-c0403f7ba3a7%3Aamzn1.symc.752cde0b-d2ce-4cce-9121-769ea438869e&pf_rd_p=340182bc-8d5c-49c7-8b69-c0403f7ba3a7&pf_rd_r=QKK8TZS001FHM8Y8KF54&pd_rd_wg=yo844&pd_rd_r=b460c84e-43d1-4aa9-9f3f-bb40bbcc9425&ref_=pd_hp_d_atf_ci_mcx_mr_&th=1&psc=1"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

### Using `soup.find()` to extract data

To extract specific information, we can use the `find()` method of the `BeautifulSoup` object. This method returns the first matching tag based on the provided criteria.

First, we need to inspect the HTML to find the relevant tags and attributes for the data we want to extract. For example, to get the product title, we might look for an `<h1>` tag with a specific ID or class.

```python
# To find the product title, we inspect the HTML. A common pattern for product titles on Amazon is an <h1> tag with id 'productTitle'
product_title_element = soup.find('span', id='productTitle')

# Extract the text from the element, strip leading/trailing whitespace
if product_title_element:
    product_title = product_title_element.get_text(strip=True)
    print(f"Product Title: {product_title}")
else:
    print("Product title not found.")
```

### Using `soup.find_all()` to extract multiple elements

The `find_all()` method is used to find all occurrences of a tag that match the given criteria. It returns a list of Tag objects. We can iterate through this list to extract information from each found element.

For customer reviews, we typically look for a container element that holds each individual review, and then extract details like the review text and rating from within each container.

### Simple `find_all` Example

Let's create a very basic HTML structure and use `find_all` to extract all paragraph (`<p>`) elements.

```python
paragraphs = soup.find_all('p')

for i, p_tag in enumerate(paragraphs):
    print(f"Paragraph {i+1}: {p_tag.get_text(strip=True)}")

print(f"\nTotal paragraphs found: {len(paragraphs)}")
```
